```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/sample.csv")
df.head()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['age'], bins=5, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Sample Data')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('outputs/figures/task01_age_histogram.png', dpi=150)
plt.show()
```

```python
counts = df['gender'].value_counts()
plt.figure(figsize=(6,4))
counts.plot(kind='bar', color=['pink','lightblue'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('outputs/figures/task01_gender_bar.png', dpi=150)
plt.show()
```

```python
# Observations
# - Most participants are between 19 and 25 years old.
# - The sample has slightly more female participants than male.
# - The distribution is roughly uniform with a small peak around 21â€“23.
```
